```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
df = pd.read_csv(r"C:\Users\mohan reddy\OneDrive\Desktop\New folder\New folder\Sample - Superstore.csv",encoding='latin1')   # or encoding='ISO-8859-1')
df.head()
```

```python
# Sales by category
df.groupby('Category')['Sales'].sum().plot(kind='bar', title="Sales by Category", color=['blue', 'orange', 'red'])
plt.ylabel("Total Sales")
plt.show()
```

```python
# Profit vs Discount
df.plot.scatter(x='Discount', y='Profit', title="Profit vs Discount")
plt.show()
```

```python
# order date vs monthly sales
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.groupby(df['Order Date'].dt.to_period("M"))['Sales'].sum().plot(title="Monthly Sales Trend")
plt.ylabel("Total sales")
plt.show()
```

```python
# sales vs category
df.groupby('Category')['Sales'].sum().plot(kind='bar', title="Sales by Category", color=['skyblue','orange','green'])
plt.ylabel("Total Sales")
plt.show()
```

```python
# profit vs region
df.groupby('Region')['Profit'].sum().plot(kind='bar', title="Profit by Region", color=['red', 'blue', 'orange', 'green', 'yellow'])
plt.ylabel("Total Profit")
plt.show()
```

```python
# sales vs trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Order Date'].dt.to_period("M"))['Sales'].sum()

monthly_sales.plot(title="Monthly Sales Trend", figsize=(12,5))
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

```python
#Top 10 Sub-Categories by Sales
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar', title="Top 10 Sub-Categories by Sales", color='purple')
plt.ylabel("Total Sales")
plt.show()
```

```python
# heatmap
pivot = df.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Profit by Region & Category")
plt.show()
```
